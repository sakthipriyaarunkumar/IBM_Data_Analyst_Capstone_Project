## Setup: Working with the Database



**Install and import the required libraries**


```python
!pip install pandas
!pip install matplotlib

import pandas as pd
import matplotlib.pyplot as plt
```

#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
## Write your code here
df['Age'] = df['Age'].replace({'Under 18 years old': 18, '35-44 years old': 40, '45-54 years old': 50, '18-24 years old': 20, '25-34 years old':30,  
                               '55-64 years old':60, 'Prefer not to say':0, '65 years or older':65}) 
```

```python
df['Age'] = df['Age'].astype('int')
```

```python
plt.figure(figsize=(15,6))
df.plot(kind='scatter', x='Age', y='JobSatPoints_6') 
plt.title('Scatter plot for Age Vs Job satisfaction') 
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
plt.figure(figsize=(15,6))
df.plot(kind='scatter', y='ConvertedCompYearly', x='JobSatPoints_6') 
plt.title('Scatter plot for ConvertedCompYearly Vs Job satisfaction') 
plt.show()
```

### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
!pip install seaborn
import seaborn as sns
```

```python
## Write your code here
plt.figure(figsize=(15,6))
sns.regplot(data=df, x='Age', y='JobSatPoints_6') 
plt.title('Regression plot for Age Vs Job satisfaction') 
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
df['YearsCodePro'] = df['YearsCodePro'].replace({'Less than 1 year':0, 'More than 50 years': 50})
df['YearsCodePro'] = df['YearsCodePro'].fillna(0)
df['YearsCodePro'] = df['YearsCodePro'].astype('int')
```

```python
## Write your code here
plt.figure(figsize=(15,6))
df.plot(kind='scatter', x='Age', y='YearsCodePro') 
plt.title('Scatter plot for Age Vs Work experience') 
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
## Write your code here
plt.figure(figsize=(15,6))
sns.scatterplot(data=df, x='JobSatPoints_6', y='ConvertedCompYearly', size='Age', alpha=0.5, edgecolors='black') 
plt.title('Bubble plot for ConvertedComp Yearly Vs Job satisfaction with size of Age') 
plt.show()
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
## Write your code here
lang = df['LanguageHaveWorkedWith'].value_counts().head().index
lang
```

```python
df1 = df.loc[df['LanguageHaveWorkedWith'].isin(lang)]
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(data=df1, x='LanguageHaveWorkedWith', y='JobSatPoints_6') 
plt.title('Scatter plot for Top 5 LanguageHaveWorkedWith Vs Job satisfaction') 
plt.show()
```

### Task 4: Scatter Plot Comparisons Across Groups


#### 1. Scatter Plot for Compensation vs. Job Satisfaction by Employment Type


Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), categorized by employment type (`Employment`). Use color coding or markers to differentiate between employment types.


```python
## Write your code here
df['Employment'] = df['Employment'].apply(lambda x: 'Employed, full-time' if x.startswith('Employed, full-time') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Employed, part-time' if x.startswith('Employed, part-time') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Not employed' if x.startswith('Not employed') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Student, full-time' if x.startswith('Student, full-time') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Student, part-time' if x.startswith('Student, part-time') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Independent contractor' if x.startswith('Independent contractor') else x)
```

```python
plt.figure(figsize=(15,6))
sns.scatterplot(data=df, x='JobSatPoints_6', y='ConvertedCompYearly', markers=True, hue='Employment')  
plt.title('Scatter plot for Age Vs Job satisfaction') 
plt.show()
```

#### 2. Scatter Plot for Work Experience vs. Age Group by Country


Compare work experience (`YearsCodePro`) across different age groups (`Age`) and countries (`Country`). Use colors to represent different countries and markers for age groups.


```python
## Write your code here
plt.figure(figsize=(11,6))
sns.scatterplot(data=df, x='Age', y='YearsCodePro', markers=True, hue='Country')  
plt.title('Scatter plot for Age Vs Job satisfaction') 
plt.legend(bbox_to_anchor=(1.1,1.05))
plt.show()
```
