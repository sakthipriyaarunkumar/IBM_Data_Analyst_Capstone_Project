### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
df['Country'].isna().sum()
```

```python
# Write your code here
df['Country'].unique()
```

```python
df['EdLevel'].value_counts()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
df['Country'] = df['Country'].fillna('Not specified') 
```

```python
## Write your code here
df['Country'] = df['Country'].replace({'United Kingdom of Great Britain and Northern Ireland': 'UK and Ireland', 'Iran, Islamic Republic of...':'Iran', 
                       'Venezuela, Bolivarian Republic of...':'Venezuela',"Democratic People's Republic of Korea":'Korea', 'Congo, Republic of the...':'Congo', 
                       'Micronesia, Federated States of...': 'Micronesia'})  
df['Country'].unique()
```

```python
df['EdLevel'] = df['EdLevel'].fillna('Something else')
df['EdLevel'].unique()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
df['Employment'].unique()
```

```python
## Write your code here
df['Employment'] = df['Employment'].apply(lambda x: 'Employed' if x.startswith('Employed') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Not employed' if x.startswith('Not employed') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Student' if x.startswith('Student') else x)
df['Employment'] = df['Employment'].apply(lambda x: 'Independent contractor' if x.startswith('Independent contractor') else x)
```

```python
df['Employment'].unique()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here
df.isnull().sum().sort_values(ascending=False)
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(df['ConvertedCompYearly'].mean())
```

```python
df['ConvertedCompYearly'].isna().sum()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode().iloc[0])
```

```python
df['RemoteWork'].isna().sum()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here
df['ConvertedCompYearly'] = (df['ConvertedCompYearly'] - df['ConvertedCompYearly'].min())/(df['ConvertedCompYearly'].max() - df['ConvertedCompYearly'].min())
```

```python
df['ConvertedCompYearly']
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here
df['ConvertedCompYearly' ] = np.log10(df['ConvertedCompYearly'])
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
df['YearsCodePro'] = df['YearsCodePro'].replace({np.nan: '0', 'Less than 1 year': '0', 'More than 50 years':'50'})
```

```python
df['YearsCodePro'] = df['YearsCodePro'].astype('int')
```

```python
groups = ['Between 0 to 10 years', 'Between 10 to 20 years', 'Between 20 to 30 years', 'Between 30 to 40 years', 'Above 40 years']
bins = np.linspace(min(df['YearsCodePro']), max(df['YearsCodePro']), 6)

df['ExperienceLevel'] = pd.cut(df['YearsCodePro'], bins, labels=groups, include_lowest=True)
```
